#tienda

```python
class Producto:
    def __init__(self, nombre, precio, cantidad):
        self.nombre = nombre
        self.precio = precio
        self.cantidad = cantidad

    def __str__(self):
        return f"Producto: {self.nombre:<15} | Precio: ${self.precio:>6.2f} | Cantidad: {self.cantidad:>3}"

    def actualizar_cantidad(self, cambio):
        """Actualiza la cantidad (puede ser positivo para añadir o negativo para restar)"""
        self.cantidad += cambio
        if self.cantidad < 0:
            self.cantidad = 0

    def actualizar_precio(self, nuevo_precio):
        if nuevo_precio >= 0:
            self.precio = nuevo_precio
        else:
            print("Error: El precio no puede ser negativo")

class Tienda:
    def __init__(self, nombre, lista_productos=None):
        self.nombre = nombre
        self.productos = []

        if lista_productos:
            for prod in lista_productos:
                self.agregar_producto(Producto(*prod))

    def agregar_producto(self, producto):
        """Agrega un producto o actualiza cantidad si ya existe"""
        existente = self.buscar_producto(producto.nombre)
        if existente:
            existente.actualizar_cantidad(producto.cantidad)
            print(f"Producto '{producto.nombre}' ya existía. Se actualizó la cantidad.")
        else:
            self.productos.append(producto)
            print(f"Producto '{producto.nombre}' agregado correctamente.")

    def mostrar_inventario(self):
        print(f"\n{' INVENTARIO ':=^40}")
        print(f"Tienda: {self.nombre}")
        print("-" * 40)
        print(f"{'Nombre':<15} | {'Precio':>10} | {'Cantidad':>8}")
        print("-" * 40)
        for producto in sorted(self.productos, key=lambda x: x.nombre):
            print(producto)
        print("=" * 40)
        total_inventario = sum(p.precio * p.cantidad for p in self.productos)
        print(f"Valor total del inventario: ${total_inventario:.2f}\n")

    def buscar_producto(self, nombre):
        for producto in self.productos:
            if producto.nombre.lower() == nombre.lower():
                return producto
        return None

    def vender_producto(self, nombre, cantidad):
        producto = self.buscar_producto(nombre)
        if not producto:
            print(f"Error: Producto '{nombre}' no encontrado.")
            return False

        if cantidad <= 0:
            print("Error: La cantidad debe ser mayor que cero")
            return False

        if producto.cantidad < cantidad:
            print(f"Error: Stock insuficiente. Hay {producto.cantidad} unidades de {producto.nombre}")
            return False

        producto.actualizar_cantidad(-cantidad)
        total = producto.precio * cantidad
        print(f"Venta exitosa: {cantidad} x {producto.nombre} = ${total:.2f}")
        return True

# Lista de productos predefinidos (nombre, precio, cantidad)
lista_productos = [
    ("Leche entera", 2.50, 30),
    ("Pan integral", 1.80, 50),
    ("Huevos blancos", 3.75, 25),
    ("Arroz blanco", 4.30, 40),
    ("Aceite vegetal", 5.20, 20),
    ("Café molido", 7.90, 15),
    ("Azúcar blanca", 3.10, 35)
]

if __name__ == "__main__":
    # Crear tienda con la lista de productos
    tienda = Tienda("Supermercado Python", lista_productos)

    # Mostrar inventario inicial
    tienda.mostrar_inventario()

    # Realizar algunas operaciones
    tienda.vender_producto("Leche entera", 5)
    tienda.vender_producto("Pan integral", 10)
    tienda.vender_producto("Producto inexistente", 5)  # Debe fallar

    # Agregar un nuevo producto
    tienda.agregar_producto(Producto("Chocolate", 4.50, 25))

    # Actualizar precio de un producto
    producto = tienda.buscar_producto("huevos blancos")
    if producto:
        producto.actualizar_precio(4.00)

    # Mostrar inventario actualizado
    tienda.mostrar_inventario()
```
