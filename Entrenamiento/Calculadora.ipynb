```python
import math

# Funciones aritméticas básicas
def suma(a, b):
    return a + b

def resta(a, b):
    return a - b

def multiplicacion(a, b):
    return a * b

def division(a, b):
    if b == 0:
        return "Error: División por cero"
    return a / b

def potencia(a, b):
    return math.pow(a, b)

def raiz_cuadrada(a):
    if a < 0:
        return "Error: Raíz de negativo"
    return math.sqrt(a)

def raiz_nesima(a, n):
    if a < 0 and n % 2 == 0:
        return "Error: Raíz par de negativo"
    if n == 0:
        return "Error: Índice cero"
    return math.pow(a, 1/n)

# Funciones trigonométricas (en grados por defecto)
def seno(angulo, grados=True):
    if grados:
        angulo = math.radians(angulo)
    return math.sin(angulo)

def coseno(angulo, grados=True):
    if grados:
        angulo = math.radians(angulo)
    return math.cos(angulo)

def tangente(angulo, grados=True):
    if grados:
        angulo = math.radians(angulo)
    return math.tan(angulo)

# Funciones trigonométricas inversas (devuelven grados por defecto)
def arcoseno(valor, grados=True):
    if valor < -1 or valor > 1:
        return "Error: Valor fuera de rango [-1,1]"
    resultado = math.asin(valor)
    return math.degrees(resultado) if grados else resultado

def arcocoseno(valor, grados=True):
    if valor < -1 or valor > 1:
        return "Error: Valor fuera de rango [-1,1]"
    resultado = math.acos(valor)
    return math.degrees(resultado) if grados else resultado

def arcotangente(valor, grados=True):
    resultado = math.atan(valor)
    return math.degrees(resultado) if grados else resultado

    # Funciones logarítmicas
def logaritmo_natural(x):
    if x <= 0:
        return "Error: Log de número ≤ 0"
    return math.log(x)

def logaritmo_base10(x):
    if x <= 0:
        return "Error: Log de número ≤ 0"
    return math.log10(x)

def logaritmo_base(x, base):
    if x <= 0 or base <= 0 or base == 1:
        return "Error: Base inválida o número ≤ 0"
    return math.log(x, base)


def mostrar_menu():
    print("\n=== CALCULADORA ===")
    print("1. Operaciones básicas")
    print("2. Potencias y raíces")
    print("3. Funciones trigonométricas")
    print("4. Funciones trigonométricas inversas")
    print("5. Funciones logarítmicas")
    print("6. Salir")

def menu_basicas():
    print("\nOperaciones básicas:")
    print("1. Suma")
    print("2. Resta")
    print("3. Multiplicación")
    print("4. División")
    print("5. Volver")

def menu_potencias():
    print("\nPotencias y raíces:")
    print("1. Potencia")
    print("2. Raíz cuadrada")
    print("3. Raíz n-ésima")
    print("4. Volver")

def menu_trigonometricas():
    print("\nFunciones trigonométricas:")
    print("1. Seno")
    print("2. Coseno")
    print("3. Tangente")
    print("4. Volver")

def menu_inversas_trig():
    print("\nFunciones trig inversas:")
    print("1. Arcoseno")
    print("2. Arcocoseno")
    print("3. Arcotangente")
    print("4. Volver")

def menu_logaritmos():
    print("\nFunciones logarítmicas:")
    print("1. Logaritmo natural (ln)")
    print("2. Logaritmo base 10 (log10)")
    print("3. Logaritmo con base personalizada")
    print("4. Volver")

def main():
    while True:
        mostrar_menu()
        opcion = input("Seleccione una categoría (1-6): ")

        if opcion == '6':
            print("¡Gracias por usar la calculadora!")
            break

        elif opcion == '1':  # Operaciones básicas
            while True:
                menu_basicas()
                sub_op = input("Seleccione operación (1-5): ")

                if sub_op == '5':
                    break

                try:
                    if sub_op in ('1', '2', '3', '4'):
                        a = float(input("Primer número: "))
                        b = float(input("Segundo número: "))

                        if sub_op == '1':
                            print(f"Resultado: {suma(a, b):.0f}")
                        elif sub_op == '2':
                            print(f"Resultado: {resta(a, b):.0f}")
                        elif sub_op == '3':
                            print(f"Resultado: {multiplicacion(a, b):.0f}")
                        elif sub_op == '4':
                            res = division(a, b)
                            print(f"Resultado: {res:.2f}" if isinstance(res, float) else res)
                except ValueError:
                    print("Error: Ingrese números válidos")

        elif opcion == '2':  # Potencias y raíces
            while True:
                menu_potencias()
                sub_op = input("Seleccione operación (1-4): ")

                if sub_op == '4':
                    break

                try:
                    if sub_op == '1':
                        a = float(input("Base: "))
                        b = float(input("Exponente: "))
                        print(f"Resultado: {potencia(a, b):.0f}")
                    elif sub_op == '2':
                        a = float(input("Número: "))
                        res = raiz_cuadrada(a)
                        print(f"Resultado: {res:.2f}" if isinstance(res, float) else res)
                    elif sub_op == '3':
                        a = float(input("Radicando: "))
                        n = float(input("Índice: "))
                        res = raiz_nesima(a, n)
                        print(f"Resultado: {res:.2f}" if isinstance(res, float) else res)
                except ValueError:
                    print("Error: Ingrese números válidos")

        elif opcion == '3':  # Funciones trigonométricas
            while True:
                menu_trigonometricas()
                sub_op = input("Seleccione función (1-4): ")

                if sub_op == '4':
                    break

                try:
                    ang = float(input("Ángulo: "))
                    unidad = input("¿Está en grados? (s/n): ").lower() == 's'

                    if sub_op == '1':
                        print(f"Seno: {seno(ang, unidad):.2f}")
                    elif sub_op == '2':
                        print(f"Coseno: {coseno(ang, unidad):.2f}")
                    elif sub_op == '3':
                        res = tangente(ang, unidad)
                        # Verificar si la tangente está definida
                        if abs(res) > 1e10:  # Tangente tiende a infinito
                            print("Resultado: Indefinido (coseno ≈ 0)")
                        else:
                            print(f"Tangente: {res:.2f}")
                except ValueError:
                    print("Error: Ingrese un ángulo válido")

        elif opcion == '4':  # Funciones trigonométricas inversas
            while True:
                menu_inversas_trig()
                sub_op = input("Seleccione función (1-4): ")

                if sub_op == '4':
                    break

                try:
                    val = float(input("Valor entre -1 y 1 (para arcoseno/arcocoseno): " if sub_op in ('1', '2')
                                else "Valor para arcotangente: "))
                    unidad = input("¿Mostrar resultado en grados? (s/n): ").lower() == 's'

                    if sub_op == '1':
                        res = arcoseno(val, unidad)
                        print(f"Arcoseno: {res:.2f}" if isinstance(res, float) else res)
                    elif sub_op == '2':
                        res = arcocoseno(val, unidad)
                        print(f"Arcocoseno: {res:.2f}" if isinstance(res, float) else res)
                    elif sub_op == '3':
                        print(f"Arcotangente: {arcotangente(val, unidad):.2f}")
                except ValueError:
                    print("Error: Ingrese un valor válido")
        elif opcion == '5':  # Funciones logarítmicas
            while True:
                menu_logaritmos()
                sub_op = input("Seleccione función (1-4): ")

                if sub_op == '4':
                    break

                try:
                    if sub_op == '1':
                        x = float(input("Ingrese el número (> 0): "))
                        res = logaritmo_natural(x)
                        print(f"ln({x}) = {res:.2f}" if isinstance(res, float) else res)
                    elif sub_op == '2':
                        x = float(input("Ingrese el número (> 0): "))
                        res = logaritmo_base10(x)
                        print(f"log10({x}) = {res:.2f}" if isinstance(res, float) else res)
                    elif sub_op == '3':
                        x = float(input("Ingrese el número (> 0): "))
                        base = float(input("Ingrese la base (>0 y ≠1): "))
                        res = logaritmo_base(x, base)
                        print(f"log{base}({x}) = {res:.2f}" if isinstance(res, float) else res)
                except ValueError:
                    print("Error: Ingrese un número válido")

        else:
            print("Opción no válida. Intente nuevamente.")

if __name__ == "__main__":
    main()
```
